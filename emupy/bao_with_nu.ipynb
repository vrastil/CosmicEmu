```python
import numpy as np
import pylab as plt
%matplotlib inline
```

```python
import power_spec
from emupy.structs import CosmoParam
```

```python
cosmo = CosmoParam()
# default:
# Omega_m=0.3075, Omega_b=0.0486, sigma_8=0.8159, h=0.67, n_s=0.9667, w_0=-1, w_a=0., Omega_nu=0.

z = 1.0
```

```python
pk_tot_0, k = power_spec.init_power_spectrum_tot(cosmo, z)
pk_cb_0 = power_spec.init_power_spectrum_cb(cosmo, z)[0]
```

```python
plt.figure(figsize=(15, 8))
for Omega_nu in np.linspace(0, 0.02, num=5):
    cosmo = CosmoParam(Omega_nu=Omega_nu)
    pk_cb = power_spec.init_power_spectrum_cb(cosmo, z)[0]
    plt.plot(k, pk_cb/pk_cb_0, label=r'$\Omega_\nu=%.3f$' % Omega_nu)

plt.xscale('log')
#plt.yscale('log')
plt.xlabel(r'$k$  [1/Mpc]',fontsize=22)
plt.ylabel(r'$P(k, \Omega_\nu)/P(k, \Omega_\nu=0)$',fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title("Dark matter power spectrum, z = %.1f" % z, size=30)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for Omega_nu in np.linspace(0, 0.02, num=5):
    cosmo = CosmoParam(Omega_nu=Omega_nu)
    pk_tot = power_spec.init_power_spectrum_tot(cosmo, z)[0]
    plt.plot(k, pk_tot/pk_tot_0, label=r'$\Omega_\nu=%.3f$' % Omega_nu)

plt.xscale('log')
#plt.yscale('log')
plt.xlabel(r'$k$  [1/Mpc]',fontsize=22)
plt.ylabel(r'$P(k, \Omega_\nu)/P(k, \Omega_\nu=0)$',fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title("Total matter power spectrum, z = %.1f" % z, size=30)
plt.legend()
plt.show()
```
